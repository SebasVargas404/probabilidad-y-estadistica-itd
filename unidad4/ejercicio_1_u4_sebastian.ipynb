```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
media = 70 # Media
desv_std = 4 # Desviacin estndar
num_datos = 500
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] # Mostrar los primeros 50
```

```python
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desv_std **2))
```

```python
pdf.size, num_datos
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b') # Histograma de los datos
plt.title('Distribucin Normal ')
plt.xlabel('Peso en Kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media,2), np.round(desv_std, 2)))
plt.title('Funcin de Densidad de Probabilidad (PDF) - Distrubucin Normal')
plt.xlabel('Peso en Kg de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b') # Histograma de los datos
plt.plot( x, pdf, 'k', linewidth=2, )
plt.title('Distribucin Normal ')
plt.xlabel('Peso en Kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))

# Parmetros de la distribucin normal
media = 70 # Media
desv_std = 4 # Desviacin estndar
total_estudiantes = 500

# VARIABLE para clculo de la probabilidad de que un estudiantes que pese menos de 75 kg
peso_limite = 75

# Primero necesitamos calcular la probabilidad de que un estudiante pese menos
# de 75 kg utilizando la funcin de distribucin acomulativa (CDF) de la distribucipon normal
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)
```

```python
# Clculo del nmero de estudiantes que pesan menos de 75 kg
estudiantes_menos_75kg = prob_menos_75kg * total_estudiantes

print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("Nmero de estudiantes que pesan menos de 75 kg:", estudiantes_menos_75kg)

# Graficar la distribucin normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='blue', alpha=0.5)
plt.xlabel('Peso en Kg de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.title('Probabilidad de que un estudiante pese menos de 75 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.grid(True)
plt.show()
```

```python
# Parmetros
media = 70
desv_std = 4
total_estudiantes = 500

# Lmites de inters
peso_min = 68
peso_max = 72

# Dominio para la grfica
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

# Calcular la probabilidad de estar entre 68 y 72 kg
prob_entre = norm.cdf(peso_max, media, desv_std) - norm.cdf(peso_min, media, desv_std)
estudiantes_entre = prob_entre * total_estudiantes

# Graficar la distribucin y el rea de inters
plt.figure(figsize=(6, 4))
plt.plot(x, y, 'k', linewidth=2, label='Distribucin Normal')
plt.fill_between(x, y, where=(x >= peso_min) & (x <= peso_max), color='blue', alpha=0.5,
                 label='Entre 68 y 72 kg')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribucin Normal: Estudiantes entre 68 y 72 kg')
plt.legend()
plt.grid(True)
plt.show()

print("Probabilidad entre 68 y 72 kg:", prob_entre)
print("Nmero aproximado de estudiantes entre 68 y 72 kg:", estudiantes_entre)
```
