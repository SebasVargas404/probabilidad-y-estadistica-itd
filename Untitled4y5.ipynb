```python
import random
```

```python
edades = []
for i in range(100):
  numero_aleatorio = random.randint(18, 70)
  edades.append(numero_aleatorio)
```

```python
  print(edades)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(edades, columns=['edades'])
df.head()
```

```python
tabla_frecuencias = df['edades'].value_counts().reset_index()
tabla_frecuencias.columns = ['edades', 'frecuencia']
tabla_frecuencias =  tabla_frecuencias.sort_values(by='edades')
print("tabla de distribuicion de frecuencias: ")
tabla_frecuencias
```

```python
tabla_frecuencias['frecc relativa'] = tabla_frecuencias['frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['frecuencia porcentual'] = tabla_frecuencias['frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['frecuencia acumulada'] = tabla_frecuencias['frecuencia'].cumsum()
tabla_frecuencias
```

```python
df = pd.DataFrame({'edad' : edades})
df
```

```python
n = len(edades)
k = int(1 + math.log2(100))
print("valor de n = ", n, " \nNumeros de intervalos =", k)
```

```python
tabla_frecuencias = pd.cut(df['edad'], bins=k, include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns = ['clase', 'frec']
tabla_frecuencias
```

```python
tabla_frecuencias = tabla_frecuencias.columns.sort_values(by='clase')
tabla_frecuencias
```
