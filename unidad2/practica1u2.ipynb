```python
import math as mt
from itertools import combinations
from itertools import combinations_with_replacement
import numpy as np
```

```python
nombres  = ["Cesar", "Francisco", "Diego", "Laura", "Luci", "Luis", "Vale", "Alberto", "Jose", "Daniel"]
n = len(nombres)
r = 2
cb = mt.factorial(n) / (mt.factorial(r) * mt.factorial(n - r))
print("Existen ", cb, " posibles combinaciones sin repeticin del total  de ", n, " nombres ", " en grupos de", r, " en ", r)
print("Aqui las combinaciones")
combinacionesNombres = list(combinations(nombres, r))
for combinacion in combinacionesNombres:
    print(combinacion)
```
